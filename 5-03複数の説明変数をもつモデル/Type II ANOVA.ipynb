```python
variable = ""

for i in range(0, len(names)):
    
    if i == 0:
        variable += str(names[i])
    else:
        variable += "+"+str(names[i])
```

```python
formula = ""
formula = "sales~"+variable
```

```python
formula
```

```python
# モデルを作成
lm_sales2 = smf.ols(formula, data=sales_model).fit()
```

```python
sm.stats.anova_lm(lm_sales2, typ=2).round(3)
```
