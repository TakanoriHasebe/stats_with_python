そこで登場するのが、一般化線形モデル（GLM）。<br>
一般化線形モデルを使うことで、分類の問題も回帰の問題も統一的に扱うことができるようになった。<br>

### 一般化線形モデルの構成要素

一般化線形モデルは以下の３つを構成要素としている。<br>
<ol>
    <li>母集団の従う確率分布</li>
    <li>線形予測子</li>
    <li>リンク関数</li>
</ol>
構成要素が３つあるということは、これらの要素をデータに合わせて柔軟に変化可能ということ。<br>

### 確率分布

一般化線形モデルは正規分布や二項分布、ポアソン分布などに適用できる。<br>

### 線形予測子

線形予測子とは、説明変数を線形の関係式で表現したもの。<br>
例えば、ビールの売上という応答変数を、気温という説明変数から予測する場合は以下のようになる。<br>

$\beta_{0}+\beta_{1}×気温(℃)$

ここで、この線形予測子をそのまま用いるのには問題が生じる。<br>

### リンク関数

リンク関数は応答変数と線形予測子の対応を取るために使用される。<br>
応答変数にリンク関数を適用する。<br>

ここで、ビールの販売個数を以下の式で考えるとする。<br>
$ビールの販売個数 = \beta_{0} + \beta_{1}×気温（℃）$

ビールの販売個数がマイナスになることは無いが、気温や回帰係数の設定によってはマイナスになる。<br>
そこで、リンク関数の出番になる。<br>
$log[ビールの販売個数] = \beta_{0}+\beta_{1}×気温（℃）$

両辺のexpをとって<br>
$ビールの販売個数 = exp[\beta_{0} + \beta_{1}×気温(℃)]$

上記のように応答変数にリンク関数を適用することで、<br>
0以上のカウントデータや、[0, 1]の範囲をとる成功確率といったものを対象にとって予測することが可能。<br>
