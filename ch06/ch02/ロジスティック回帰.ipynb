ロジスティック回帰は、確率分布に二項分布を用い、<br>
リンク関数にロジット関数を用いた一般化線形モデルのこと。<br>
説明変数は複数あっても構わないし、連続型・カテゴリ型が混在していても支障がない。<br>

### この章の例題

テストの合格・不合格を予測することについて考える。<br>
勉強時間によって、テストの合否が変わる、という構造で数理モデルにすると、線形予測子は以下のようになる。<br>
$\beta_{0} + \beta_{1}×勉強時間$

### 二値判別の問題

応答変数を「合格ならば１、不合格ならば０をとる２値確率変数」だと考える。<br>
この時テストの合否を以下のように予測するのはするのは明らかに過ち。<br>
$テストの合否(合格1, 不合格0) = \beta_{0}+\beta_{1}×勉強時間$<br>
なぜなら、勉強時間は連続型の変数なので、予測値であるテストの合否が小数点以下の値を取ることもある。<br>

テストの合格率を勉強時間で説明する数理モデルを考える。<br>
$テストの合格率 = \beta_{0}+\beta_{1}×勉強時間$<br>
ここでまだ問題が存在する。<br>
合格率はその定義上[0, 1]の範囲をとるべきだからである。<br>
この問題はリンク関数としてロジット関数をとることで大丈夫になる。<br>

### ロジスティック回帰の構造

ロジスティック回帰とは、確率分布に二項分布を用い、<br>
リンク関数にロジット回帰を用いた一般線形モデルのこと。<br>

リンク関数にロジット関数を使うと、テストの合格率と勉強時間の関係は以下のようになる。<br>

$log(p / 1-p) = \beta_{0} + \beta_{1}×勉強時間$

両辺にロジスティック関数を適用すると、<br>
以下のように変更可能。<br>

$p = 1 / (1 + exp[-(\beta_{0} + \beta_{1}×勉強時間)])$

勉強時間が５時間だった生徒が10人いたとする。<br>
この時の合格者人数Mは、成功確率が上記で、試行回数が１０の二項分布に従うと想定される。<br>

### 二項分布

```python
# DataVisualization
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

# Stats
import scipy as sp
from scipy import stats
import statsmodels.formula.api as smf
import statsmodels.api as sm
# 表示桁数の指定
%precision 3
```

```python
# 表が出る確率が50%であるコインを２回投げて、そのうち１回が表である確率について
sp.stats.binom.pmf(k=1, n=2, p=0.5)
```

```python
# N = 10, p = 0.2の二項分布
binomial = sp.stats.binom(n=10, p=0.2)
# 乱数
np.random.seed(1)
rvs_binomial = binomial.rvs(size=10000)
# 確率質量関数
m = np.arange(0, 10, 1)
pmf_binomial = binomial.pmf(k=m)
# 乱数のヒストグラムと確率質量関数を重ねる
sns.distplot(rvs_binomial, bins=m, kde=False, norm_hist=True, color='gray')
plt.plot(m, pmf_binomial, color='black')
```
